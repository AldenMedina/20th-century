```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create DataFrame from tags
df_tags = pd.DataFrame(tags_list, columns=["Word", "POS"])

# Count top 10 POS tags
top_tags = df_tags['POS'].value_counts().nlargest(10).reset_index()
top_tags.columns = ['POS', 'Count']

# Plot it
plt.figure(figsize=(10,5))
sns.barplot(x='POS', y='Count', data=top_tags)
plt.title("Top 10 POS Tags")
plt.xlabel("Part of Speech")
plt.ylabel("Count")
plt.show()
```

Nouns and proper nouns dominate the text, which makes sense since the article discusses events, people, and countries from the 20th century. There are also many verbs and adjectives indicating action and description.


```python
# Define POS tag categories
pos_categories = {
    'Noun': ['NN', 'NNS', 'NNP', 'NNPS'],
    'Verb': ['VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ'],
    'Adjective': ['JJ', 'JJR', 'JJS'],
    'Adverb': ['RB', 'RBR', 'RBS']
}

# Count POS
pos_counts = df_tags['POS'].value_counts()

# Aggregate counts by category
category_counts = {}
for category, tags in pos_categories.items():
    category_counts[category] = sum(pos_counts.get(tag, 0) for tag in tags)

# Create bar chart
plt.figure(figsize=(8,5))
sns.barplot(x=list(category_counts.keys()), y=list(category_counts.values()))
plt.title("POS Tag Categories")
plt.ylabel("Count")
plt.xlabel("Category")
plt.show()
```

```python
import pandas as pd

# Ensure you have your tags list from TextBlob
df_text = pd.DataFrame(tags_list, columns=['Word', 'Word type'])

# Filter for nouns
df_nouns = df_text[df_text['Word type'].isin(["NN", "NNS", "NNP", "NNPS"])]
df_nouns = df_nouns.groupby('Word').size().reset_index(name='Count')

# Filter out irrelevant tokens (months, brackets, etc.)
months = ['January', 'February', 'March', 'April', 'May', 'June',
          'July', 'August', 'September', 'October', 'November', 'December']
irrelevant = months + [']', '[', 'edit']

df_nouns_clean = df_nouns[~df_nouns['Word'].isin(irrelevant)]

# Plot top 15 nouns
import seaborn as sns
import matplotlib.pyplot as plt

top_nouns = df_nouns_clean.sort_values(by='Count', ascending=False).head(15)
plt.figure(figsize=(12,6))
sns.barplot(x='Word', y='Count', data=top_nouns)
plt.title("Top 15 Cleaned Nouns")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Verbs
df_verbs = df_text[df_text['Word type'].str.startswith('VB')]
df_verbs = df_verbs.groupby('Word').size().reset_index(name='Count')
top_verbs = df_verbs.sort_values(by='Count', ascending=False).head(15)

plt.figure(figsize=(12,6))
sns.barplot(x='Word', y='Count', data=top_verbs)
plt.title("Top 15 Verbs")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Adjectives
df_adjs = df_text[df_text['Word type'] == 'JJ']
df_adjs = df_adjs.groupby('Word').size().reset_index(name='Count')
top_adjs = df_adjs.sort_values(by='Count', ascending=False).head(15)

plt.figure(figsize=(12,6))
sns.barplot(x='Word', y='Count', data=top_adjs)
plt.title("Top 15 Adjectives")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Many common verbs describe actions like "was", "said", or "took". The adjectives like "great", "first", or "major" reflect the dramatic and historical nature of the content.
